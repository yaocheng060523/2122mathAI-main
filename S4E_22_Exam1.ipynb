```python
import time
from selenium import webdriver #open googledriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.support.ui import Select

PATH = "/Users/billyyu/Desktop/MathAI/WebCrawler/chromedriver"
driver = webdriver.Chrome(PATH)

link = "http://www.macaodaily.com/html/2021-11/10/node_2.htm"
driver.get(link)

topic = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="Layer1"]/table[1]/tbody/tr[2]/td/a[3]'))
)
topic.click()

    

titles = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.ID, "all_article_list"))
)
#titles = driver.find_elements_by_id("all_article_list")
for title in titles:
    print(title.text)

txt = title

x = txt.split("/")

print(x)
    
 
```
