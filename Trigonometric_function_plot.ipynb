# Import all the library we need

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interact_manual #下午先講
```

# Review the function we plotted

```python
x = np.linspace(-2*np.pi, 2*np.pi, 10000)
    
def drawsine(A = 1, omega = 1, phi = 0):   
    y = A * np.sin(omega * x + phi) 
    plt.plot(x, y, lw=2, label = "$\sin$") #lw 粗 ; label右上角
    plt.xticks(list(np.arange(-2*np.pi , 2*np.pi+1, np.pi/2)),
              ['$-2\pi$', '$-3\pi/2$', '$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']); #x軸的值和顯示
    plt.grid(color='k', linestyle='--', alpha = 0.3, linewidth=1)
    plt.legend(loc = 1) #label 定位
    plt.show()
    
def drawsine_COS_ABS(A = 1, omega = 1, phi = 0):   
    y = abs(A * np.cos(omega * x + phi))
    plt.plot(x, y, lw=2, label = "$\sin$") #lw 粗 ; label右上角
    plt.xticks(list(np.arange(-2*np.pi , 2*np.pi+1, np.pi/2)), #arrange 可以有小數
              ['$-2\pi$', '$-3\pi/2$', '$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']);
    plt.grid(color='k', linestyle='--', alpha = 0.3, linewidth=1)
    plt.legend(loc = 1) #label 定位
    plt.show()
```

```python
drawsine(1/3,2,0) #4 ->p16 - > 變1 ->(1)
```

```python
drawsine_COS_ABS(1,1,0)  #4 ->p16 - > 變1 ->(4)
```

```python
drawsine(3,0.5,np.pi/3)  #4 ->p16 - > 變1 ->(2)
```

```python
drawsine(1,-(1/3),np.pi/4) #4 ->p16 - > 變1 ->(3)
```

```python
drawsine(2,1/2,np.pi)
```

```python
drawsine(3,2,np.pi/3)
```

```python
interact(drawsine, A =(1.,3.),omega = (-2. , 2.), phi = (-2*np.pi , 2 *np.pi)) #. float 類teract(drawsine, A =(1.,3.),omega = (-2. , 2.), phi = (-2*np.pi , 2 *np.pi)) #. float 類人田十型
#重複時可以調值
#有得拉調值
#可以觀看變化個程
#改值後要enter
```
