# Try something special

```python
x = np.linspace(0, 10, 100)
y = np.sin(5*x)/(1 + x**2)
plt.plot(x, y, 'r',marker = 'o')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(5*x)/(1 + x**2)
plt.plot(x, y, 'r:') #r -> red , : ->點組成線
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.plot(x[y>0], y[y>0], 'wo') #o線 ->點
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y, lw=3) #粗
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], #math pi
          ['$0$', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']);
```

```python
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, label = "$\sin$")
plt.plot(x, y_cos, label = "$\cos$")
plt.legend(loc = 4) #label location
```

```python
def draws():
    x = np.linspace(0, 2*np.pi, 100)
    y_sin = np.sin(x)
    y_cos = np.cos(x)
    plt.plot(x, y_sin, label = "$\sin$") #show sin funiction(font latex)
    plt.plot(x, y_cos, label = "$\cos$")
    plt.legend()
    plt.grid(color='k', linestyle='--', alpha = 0.3, linewidth=1) #k black linestyle: --虛線 alpha:透明度 
```

```python
draws()
```
